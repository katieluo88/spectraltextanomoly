```python
# json_df = df[["premise", "hypothesis", "score", "output", "ground_truth_output", "Attacked"]]
json_df = df[["ex_id", "text", "score", "output", "ground_truth_output", "Attacked"]]
json_df = json_df.rename(columns={"Attacked": "label"})
json_df
```

```python
result = json_df.to_json(orient="records")
result
```

```python
import json
with open('../data/textfooler-distilbert-base-cased-snli.json', 'w') as fout:
    json.dump(result, fout)
```

```python
parsed = json.loads(result)
with open('../data/textfooler-distilbert-base-cased-snli.jsonl', 'w') as outfile:
    for entry in parsed:
        json.dump(entry, outfile)
        outfile.write('\n')
```
