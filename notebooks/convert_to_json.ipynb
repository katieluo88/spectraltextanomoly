```python
df[['premise', 'hypothesis']] = df.text.str.split('<SPLIT>', 1, expand=True)
df
```

```python
df['premise'] = df.premise.str.split(': ', 1, expand=True)[1]
```

```python
df['hypothesis'] = df.hypothesis.str.split(': ', 1, expand=True)[1]
```

```python
df.iloc[0]
```

```python
df['premise'] = df.premise.str.replace('[[', '', regex=False).str.replace(']]', '', regex=False)
df['hypothesis'] = df.hypothesis.str.replace('[[', '', regex=False).str.replace(']]', '', regex=False)
```

```python
df
```

```python
json_df = df[["premise", "hypothesis", "score", "output", "ground_truth_output", "Attacked"]]
json_df = json_df.rename(columns={"Attacked": "attacked"})
json_df
```

```python
result = json_df.to_json(orient="records")
result
```

```python
import json
with open('textfooler-distilbert-base-cased-snli.json', 'w') as fout:
    json.dump(result, fout)
```

```python
parsed = json.loads(result)
with open('textfooler-distilbert-base-cased-snli.jsonl', 'w') as outfile:
    for entry in parsed:
        json.dump(entry, outfile)
        outfile.write('\n')
```
